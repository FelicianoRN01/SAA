```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Lo primero que tendremos que hacer sera crear nuestro csv o descargarlo, una vez hemos hecho eso debemos cargarlo en el notebook y verificar que esta correctamente cargado.

- Con el head lo que haremos sera mostrar las 5 primeras filas de nuestro dataseet junto con sus respectos datos y columnas

```python
# Cargar el dataset desde una URL y mostrar las primeras 5 filas
df = pd.read_csv('https://raw.githubusercontent.com/ageron/data/main/lifesat/lifesat.csv')

# Mostrar las primeras 5 filas del dataframe para verificar que se ha cargado correctamente
df.head()
```

- Con el tail lo que haremos sera mostrar las 5 ultimas filas de nuestro dataseet junto con sus respectos datos y columnas

```python
# Mostrar las últimas 5 filas del dataframe para verificar los datos al final del dataset
print(df.tail())
```

- Con .info nos dara informacion sobre los datos que hay en cada columna, su tipo de dato y el uso de memoria que esta utilizando

```python
# El método info() proporciona un resumen conciso del DataFrame, incluyendo el índice, el número de columnas, el nombre de cada columna, el número de valores no nulos en cada columna, el tipo de datos de cada columna y el uso de memoria del DataFrame.
print(df.info())
```

- El método describe se utiliza para generar estadísticas descriptivas de un DataFrame. Este proporciona un resumen rápido de las características principales de los datos, como la media, la desviación estándar, los valores mínimos y máximos, y los percentiles.

    - count es el número de valores no nulos.
    - mean es la media aritmética.
    - std es la desviación estándar.
    - min es el valor mínimo.
    - 25%, 50% (mediana) y 75% son los percentiles.
    - max es el valor máximo.

```python
# El método describe() se utiliza para generar estadísticas descriptivas de un DataFrame.
print(df.describe())
```

- El metodo .isnull nos mostrara si hay valores nulos o no en nuestro dataseet, este valor te lo dara por columnas distintivamente

```python
# Esta línea de código imprime el número de valores faltantes en cada columna del DataFrame 'df'.
print(df.isnull().sum())
```

#### Para ver los datos graficamente necesitaremos utilizar matploblib, que es una biblioteca que se utiliza para visualizar los distintos datos

El código crea un gráfico de barras que muestra la satisfacción con la vida por país.

- Crea una figura
- Grafica los datos de satisfacción con la vida por país por barras
- Añade etiquetas a los ejes x e y
- Añade un título al gráfico
- Rota las etiquetas del eje x para evitar superposiciones
- Muestra el gráfico

- En la siguiente grafica vemos que la satisfacion esta bastante mas pareja y tenemos que Finlandia y Dinamarca tienen las mejores notas

```python
# Crear una nueva figura y ejes para el gráfico
fig, ax = plt.subplots()

# Crear un gráfico de barras con la satisfacción de vida por país
ax.bar(df['Country'], df['Life satisfaction'])

# Etiquetar el eje x
ax.set_xlabel('Country')
# Etiquetar el eje y
ax.set_ylabel('Life satisfaction')

# Añadir un título al gráfico
ax.set_title('Satisfacion de vida por pais')

# Rotar las etiquetas del eje x para evitar superposiciones
plt.xticks(rotation=90)

# Mostrar el gráfico
plt.show()
```

- Este codigo hace lo mismo que el anterior solo que esta vez estamos mostrando el GDP per capita o tambien conocido como PIB(Producto interno bruto)

    - Como vemos en la grafica, Estados unidos tiene una gran diferencia frente a las otras
    - Y España la tenemos en el top 19 lo cual me parece muy interesante

```python
# Crear una nueva figura y ejes para el gráfico
fig, ax = plt.subplots()

# Crear un gráfico de barras con el PIB per cápita por país
ax.bar(df['Country'], df['GDP per capita (USD)'])

# Etiquetar el eje x
ax.set_xlabel('Country')
# Etiquetar el eje y
ax.set_ylabel('GDP per capita (USD)')

# Añadir un título al gráfico
ax.set_title('GDP per Capita por País')

# Rotar las etiquetas del eje x para evitar superposiciones
plt.xticks(rotation=90)

# Mostrar el gráfico
plt.show()
```
